```python
# Erstelle ein Dictionary, das die Accountnamen den Followerzahlen zuordnet
follower_mapping = dict(zip(df_ubersicht_usa['Accountname'], df_ubersicht_usa['Follower']))

# Füge eine neue Spalte "Followerzahl" in df_usa hinzu und fülle sie mit den Werten aus dem Dictionary
df_usa['Followerzahl'] = df_usa['author'].map(follower_mapping)
```

```python
# Erstelle ein Dictionary, das die Accountnamen den Followerzahlen zuordnet
follower_mapping = dict(zip(df_ubersicht_ger['Accountname'], df_ubersicht_ger['Follower']))

# Füge eine neue Spalte "Followerzahl" in df_usa hinzu und fülle sie mit den Werten aus dem Dictionary
df_ger['Followerzahl'] = df_ger['author'].map(follower_mapping)
```

```python
df_usa['interaction_rate'] = (df_usa['num_comments'] + df_usa['num_likes']) / df_usa['Followerzahl']
```

```python
df_ger['interaction_rate'] = (df_ger['num_comments'] + df_ger['num_likes']) / df_ger['Followerzahl']
```

```python
from scipy import stats

# Unterteile den DataFrame in zwei Gruppen basierend auf der Variable "Inhalt"
gruppe_1 = df_usa[df_usa['Inhalt'] == 1]['interaction_rate']
gruppe_0 = df_usa[df_usa['Inhalt'] == 0]['interaction_rate']

# Führe den T-Test durch
t_statistic, p_value = stats.ttest_ind(gruppe_1, gruppe_0)

# Gib die Ergebnisse aus
print(f"T-Statistik: {t_statistic}")
print(f"P-Wert: {p_value}")
```
