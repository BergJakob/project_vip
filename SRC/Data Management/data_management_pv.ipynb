# Join USA

```python
# Pfad zum Hauptordner
main_folder_path = "/content/drive/MyDrive/Projekt_VIP/Data/USA"

# Liste, um alle DataFrames zu sammeln
dfs = []

# Alle CSV-Dateien im Ordner iterieren
for file in os.listdir(main_folder_path):
    if file.endswith(".csv"):
        file_path = os.path.join(main_folder_path, file)

        # Den Dateinamen ohne Endung extrahieren
        file_name = os.path.splitext(file)[0]

        # CSV-Datei lesen
        df = pd.read_csv(file_path, sep = ";")

        # Spalte mit dem Dateinamen hinzufügen
        df["SourceFile"] = file_name

        # DataFrame zur Liste hinzufügen
        dfs.append(df)

# Alle DataFrames zusammenfügen
combined_df = pd.concat(dfs, ignore_index=True)

# Kombinierte Datei speichern
output_file = "/content/drive/MyDrive/Projekt_VIP/Data/pro_vip_usa.csv"
combined_df.to_csv(output_file, index=False)

print(f"Kombinierte Datei gespeichert unter: {output_file}")
```

```python
df['sex'] = df['sex'].fillna('party')
df['state'] = df['state'].fillna('party')
df['party'] = df['party'].fillna('unknown')

# Gesamtanzahl der Zeilen für das Sample
sample_size = 5000

# Sampling nach "id_scraping", "type", "state" und "party"
sampled_df = (
    df.groupby(['id_scraping', 'type', 'state', 'party', 'acc_type', 'sex'], group_keys=False)
      .apply(lambda x: x.sample(frac=sample_size / len(df), random_state=42))
      .reset_index(drop=True)
)

print(sampled_df.columns)
```

```python
sampled_df
```

```python
# Verteilung im Original-Datensatz nach "SourceFile" und "type"
original_distribution = (
    df.groupby(['id_scraping', 'type', 'state', 'party', 'acc_type', 'sex']).size() / len(df)
)

# Verteilung im Sample nach "SourceFile" und "type"
sampled_distribution = (
    sampled_df.groupby(['id_scraping', 'type', 'state', 'party', 'acc_type', 'sex']).size() / len(sampled_df)
)

# Vergleich ausgeben
print("Original-Verteilung:\n", original_distribution)
print("Sample-Verteilung:\n", sampled_distribution)

# Überprüfen, ob die Verteilungen ähnlich sind
comparison = pd.concat([original_distribution, sampled_distribution], axis=1, keys=['Original', 'Sample'])
comparison
```

# Eva-Learn-Split

## VIP

```python
import pandas as pd
from sklearn.model_selection import StratifiedKFold

# Datensatz einlesen
df = pd.read_csv('/content/drive/MyDrive/Projekt_VIP/Data/Eva/USA/eva_usa_anno.csv', sep=";")

# Anzahl der Folds
k = 5

# Stratified K-Fold Initialisierung (nach "SourceFile" und "type" stratifiziert)
skf = StratifiedKFold(n_splits=k, shuffle=True, random_state=42)

# Sicherstellen, dass die Stratifizierung korrekt funktioniert
y = df[['SourceFile', 'type']].astype(str).agg('-'.join, axis=1)  # Kombination beider Spalten für die Stratifizierung

# K-Fold Split durchführen
for i, (train_idx, val_idx) in enumerate(skf.split(df, y)):
    train_df = df.iloc[train_idx]
    val_df = df.iloc[val_idx]

    # Speichern der CSV-Dateien für diesen Fold
    train_df.to_csv(f'/content/drive/MyDrive/Projekt_VIP/Data/Eva/USA/val_df_usa_{i+1}.csv', index=False, sep=";")
    val_df.to_csv(f'/content/drive/MyDrive/Projekt_VIP/Data/Eva/USA/learn_df_usa_{i+1}.csv', index=False, sep=";")

    print(f"Fold {i+1}: Val Size = {len(train_df)}, Learn Size = {len(val_df)}")

print("Stratifizierte K-Fold Cross-Validation abgeschlossen.")
```
